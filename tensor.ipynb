```python
# Create some tensors
import torch

t0 = torch.tensor(1000)                  # 0-D Tensor
t1 = torch.tensor([9, 8, 7, 6])          # 1-D Tensor
t2 = torch.tensor([[1, 2, 3], [7, 5, 3]])  # 2-D Tensor

print(t0)
print(t1)
print(t2)
```

```python
# set tensor data type
a = torch.ones((2, 3), dtype=torch.int16) # 2 rows 3 columns
print(a)
```

```python
 # cast to a different data type
b = a.to(dtype=torch.float32)
print(b)
```

```python
# check size
print(t1.size())
print(t2.shape)
```

```python
# check dimensionality
print(t1.ndimension())
print(t2.ndim)
```
